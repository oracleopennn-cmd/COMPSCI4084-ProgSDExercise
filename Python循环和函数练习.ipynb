```python
def display_menu():
    print("\n--- Name List Manager ---")
    print("1. Add a name")
    print("2. Change a name")
    print("3. Delete a name")
    print("4. View all names")
    print("5. Exit")

def add_name(name_list):
    name = input("Enter the name to add: ")
    name_list.append(name)
    print(f"‚úÖ '{name}' has been added.")

def change_name(name_list):
    if not name_list:
        print("‚ö†Ô∏è The list is empty. Nothing to change.")
        return
    view_names(name_list)
    try:
        index = int(input("Enter the number of the name to change: ")) - 1
        if 0 <= index < len(name_list):
            new_name = input("Enter the new name: ")
            old_name = name_list[index]
            name_list[index] = new_name
            print(f"üîÑ '{old_name}' has been changed to '{new_name}'.")
        else:
            print("‚ùå Invalid number.")
    except ValueError:
        print("‚ùå Please enter a valid number.")

def delete_name(name_list):
    if not name_list:
        print("‚ö†Ô∏è The list is empty. Nothing to delete.")
        return
    view_names(name_list)
    try:
        index = int(input("Enter the number of the name to delete: ")) - 1
        if 0 <= index < len(name_list):
            removed = name_list.pop(index)
            print(f"üóëÔ∏è '{removed}' has been deleted.")
        else:
            print("‚ùå Invalid number.")
    except ValueError:
        print("‚ùå Please enter a valid number.")

def view_names(name_list):
    if not name_list:
        print("üì≠ The list is currently empty.")
    else:
        print("\nüìã Current Names:")
        for i, name in enumerate(name_list, start=1):
            print(f"{i}. {name}")

def main():
    name_list = []
    while True:
        display_menu()
        choice = input("Choose an option (1-5): ")
        if choice == "1":
            add_name(name_list)
        elif choice == "2":
            change_name(name_list)
        elif choice == "3":
            delete_name(name_list)
        elif choice == "4":
            view_names(name_list)
        elif choice == "5":
            print("üëã Goodbye!")
            break
        else:
            print("‚ùå Invalid option. Please choose between 1 and 5.")

main()
```
